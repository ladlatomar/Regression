Q2. What are the key assumptions of Simple Linear Regression?

Answer:

Linearity: Relationship between X and Y is linear.

Independence: Observations are independent of each other.

Homoscedasticity: Constant variance of residuals.

Normality: Residuals are normally distributed.

Q6. What is the purpose of the least squares method in Simple Linear Regression?

Answer:

To find the best-fit line that minimizes the sum of squared differences between actual and predicted Y values.

Q7. How is the coefficient of determination (R²) interpreted?

Answer:

R² shows the proportion of variance in Y explained by X.

R² = 1 → perfect fit; R² = 0 → model explains nothing.

Q9. Main difference between Simple and Multiple Linear Regression

Answer:

Simple → one independent variable.

Multiple → two or more independent variables.

Q10. Key assumptions of Multiple Linear Regression

Answer:

Linearity of X and Y

Independence of errors

Homoscedasticity

Normality of residuals

No multicollinearity between independent variables

Q11. What is heteroscedasticity?

Answer:

Heteroscedasticity occurs when residuals have non-constant variance.

It can make regression coefficients unreliable.

Q12. How to improve a model with high multicollinearity?

Answer:

Remove correlated variables

Use Principal Component Analysis (PCA)

Use regularization techniques (Ridge or Lasso regression)

Q13. Common techniques for transforming categorical variables

Answer:

One-Hot Encoding

Label Encoding

Target Encoding

Q15. How can intercept interpretation differ between Simple and Multiple Regression?

Answer:

Simple → intercept = Y when X=0

Multiple → intercept = Y when all X variables = 0

Q16. Significance of slope in regression

Answer:

Slope shows how much Y changes per unit change in X.

Important for predictions and understanding relationships.

Q17. Intercept provides context in regression

Answer:

Indicates baseline or starting point of the dependent variable.

Q18. Limitations of R² as sole measure

Answer:

R² does not detect overfitting

Can increase with more variables even if they aren’t meaningful

Doesn’t indicate causation

Q19. Interpreting large standard error for a coefficient

Answer:

Suggests low confidence in the estimate

Coefficient might not be statistically significant

Q20. Identifying heteroscedasticity in residual plots

Answer:

Plot residuals vs predicted values

Non-constant spread → heteroscedasticity

Important because it violates regression assumptions

Q21. High R² but low adjusted R²

Answer:

Indicates many variables added that don’t contribute significantly

Adjusted R² penalizes unnecessary variables

Q22. Importance of scaling variables

Answer:

Helps improve model stability and interpretability

Especially important for regularization and gradient-based methods
